#importing libraries

```python
import cv2
from google.colab.patches import cv2_imshow
```

#sample image
reading the image with opencv library

```python
image = cv2.imread("/content/oil-painting-conceptual-abstract-picture-260nw-.jpg")
cv2_imshow(image)
```

#converting to grey image
image coverts into black and white photo

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

#converting to inverted image
negative image of grayscale image

```python
image_invert = cv2.bitwise_not(gray_image)
cv2_imshow(image_invert)
```


#converting to blur image
blurred image of previous image

```python
image_blur = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)

cv2_imshow(image_blur)
```


#finally converting to pencil sketch

```python
def final(x, y):
    return cv2.divide(x, 255 - y, scale=256)

final_image = final(gray_image, image_blur)

cv2_imshow(final_image)
```
