#splitting the data
**The train set contains the data which will be fed into the model.**

**The test set contains the data on which we test the trained and validated model.**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```


# logistic regression 

**Logistic regression is a classification algorithm used to predict the probability of a target variable(used for solving the classification problems).**



```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python

lr.fit(x_train, y_train)
```

```python
predict_lr = lr.predict(x_train)
```

```python
# metric to get performance score
print("Accuracy: ",accuracy_score(predict_lr, y_train) * 100)
```

# knn ( k-nearest neighbor )
**K-NN algorithm can be used to solve both classification and regression problems. It stores all the available data and classifies a new data case based on the similarity.**

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()

print(std)
```

```python
x_train_std = std.fit_transform(x_train)
x_test_std = std.fit_transform(x_test)


'''
print:

x_train_std

x_test_std
'''
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train_std,y_train)
```

```python
# metric to get performance score

predict_knn = knn.predict(x_test_std)

print("Accuracy: ",accuracy_score(y_test,predict_knn)*100)
```

# decision tree
**Decision Trees can be used for both classification and Regression problems. It is a tree-structured classifier, where decision nodes are used to make decisions and leaf nodes represents the outcome.**


```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
predict_dt = dt.predict(x_test)
```

```python
# metric to get performance score

print("Accuracy: ",accuracy_score(predict_dt, y_test)*100)
```
